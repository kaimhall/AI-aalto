### Exercise miscellaneous-math-quiz-3

```python
# solve argmax with sympy

x = Symbol('x', real=True)
f = (-x**2) - (2 * x)
derivative = f.diff(x)
print(f'argmax of f = {solve(derivative, x)}')
```

### Basic-logic-quiz-1

```python
# sympy variables
from sympy.abc import A, B, C
from sympy import And, Or, Not
```

```python
# {a=0, b=1, c=0}

A = False
B = True
C = False

exp1 = Or(Not(A),B)
exp2 = Or(A, C)
exp3 = And(exp1, exp2)
exp3
```

```python
# {a=1, b=0, c=0}

A = True
B = False
C = False

exp1 = Or(Not(A),B)
exp2 = Or(A, C)
exp3 = And(exp1, exp2)
exp3
```

```python
# {a=1, b=1, c=1}

A = True
B = True
C = True

exp1 = Or(Not(A),B)
exp2 = Or(A, C)
exp3 = And(exp1, exp2)
exp3
```

```python
# {a=, b=0, c=} => True

A = False
B = False
C = False

exp1 = Or(Not(A),B)
exp2 = Or(A, C)
exp3 = And(exp1, exp2)
exp3
```

```python
from sympy.logic.boolalg import truth_table
from sympy.abc import a, b

table = truth_table((a & b), [a, b])
for t in table:
    print('{0} -> {1}'.format(*t))
```

### Exercise basic-logic-quiz-4

```python
# there exists X for all YX such that YX = Y + 1 

x = 1
y = 0
yx = x * y

yx == y + 1
```

```python
# for all X there exists Y such that X = Y + 1
x = 0
y = -1

x == y + 1
```

```python
# for all X and some Y, X = 2Y
x = 3
y = 1
x == 2 * y
```

```python
# for all X and some Y, 2X = Y
x = -1
y = -2

2*x == y
```

```python
A = set([1,2,3,4,5])
B = set([2,3,4,6])
C = set([3,5,6])

D = A.intersection(B.union(C))
E = A.intersection(B).union(A.intersection(C))
E == D
```
