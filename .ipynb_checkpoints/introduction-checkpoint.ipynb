3.1<br>
describe value of each of the rest of the outputs o1; o2; o3; c as a propositional formula that only has occurrences of the
atomic propositions i0; i1; i2; i3 that describe the values of the input bits

```python
# output values as propositional formula 3.1

atoms = ['i0','i1','i2','i3']
out0 = '~i0'
out1 = 'i1 xor i0'
out2 = 'i2 xor (i0 and i1)'
out3 = 'i3 xor (i0 and i1 and i2)'
c = 'i0 and i1 and i2 and i3'
out = [out0,out1,out2, out3, c]
out
```

```python
table = ttg.Truths(atoms,out).as_pandas()
table.set_axis(['i0', 'i1', 'i2', 'i3', 'o0', 'o1', 'o2',
       'o3', 'c'], axis=1, inplace=True)
```

### Satisfiability

<br>


<b>Satisfiability:</b>

<br>
model is a synonym for a <b>satisfying valuation (=T)</b> .<br>Existence of a model (consistency) is perhaps the most fundamental question in logic and the essential requirement behind satisfiability<br>
<br>
<b>Formula is unsatisfiable</b>, if it has no models. A formula is unsatisfiable if F occurs everywhere in its column.<br>
<b>Set of formulas is unsatisfiable</b> if for every row of the joint truth table F occurs in the column of at least one formula

### exercise 7.1

1. If a set is satisfiable, then any of its subsets is also satisfiable.

```python
atoms = ['A','B','C']
f0 = '(A and B)'
f1 = '(B or C)'
f2 = '(A and (B or C))'
f3 = f'{f0} and {f1} and {f2}' 
phi = [f0,f1,f2,f3]

table = ttg.Truths(atoms, phi, ints=False).as_pandas()
table

# 
```

1. if subset is not satisfiable, set is not satisfiable

```python
atoms = ['A','B','C']
f0 = '(A and B)'
f1 = '(B or C)'
f2 = '(A or B) and (A => ~C) and (B => ~C) and C'

f3 = f'{f0} and {f1} and {f2}' 
phi = [f0,f1,f2,f3]

table = ttg.Truths(atoms, phi, ints=False).as_pandas()
table.columns = ['A', 'B', 'C', 'phi0','phi1','phi2', 'psi']
table

# if subset is not satisfiable set is not satisfiable 
```

if subset is not satisfiable, set is not satisfiable. Therefore for set to be satisfiable, subsets must be satisfiable.

### exercise 7.2
<br>
If phi1 and phi2 are satisfiable then phi1 AND phi2 is satisfiable.

```python
atoms = ['A','B','C']
phi1 = 'A'
phi2 = '~A'

psi = f'{phi1} and {phi2}' 
phi0 = [phi1,phi2,psi]

table = ttg.Truths(atoms, phi0, ints=False).as_pandas()
table.columns = ['A', 'B', 'C', 'phi1','phi2', 'psi']
table

# if both phi are satisfiable, AND is satisfiable
# if both phi are satisfiable, OR is satisfiable
```

If phi1 is not, and phi2 is satisfiable then phi1 AND phi2 is satisfiable.

```python
atoms = ['A','B','C']

phi1 = '(A or B) and (A implies ~C) and (B => ~C) and C'
phi2 = '(B implies C)'

psi = f'{phi1} or {phi2}' 
phi0 = [phi1,phi2,psi]

table = ttg.Truths(atoms, phi0, ints=False).as_pandas()
table.columns = ['A', 'B', 'C', 'phi1','phi2', 'psi']
table

# if phi1 is not satisfiable AND is not satisfiable
# if phi1 is not satisfiable OR is satisfiable
```

when phi1 is not satisfiable AND becomes not satisfiable. therefore both phi have to be satisfiable for AND to be satisfiable

```python
atoms = ['i0','i1','i2','i3','a1','a2','a3','o0','o1','o2','o3','c']
a1 = '(a1 = (i0 and i1))'
a2 = '(a2 = (a1 and i2))'
a3 = '(a3 = (a2 and i3))'
o0 = '(o0 = ~i0)'
o1 = '(o1 = (i1 xor i0))'
o2 = '(o2 = (i2 xor a1))'
o3 = '(o3 = (i3 xor a2))'
c =  '(c = a3)'

phi = f'{a1} and {a2} and {a3} and {o0} and {o1} and {o2} and {o3} and {c}'
phi1 = f'{phi} and i0 and o1'
phi2 = f'{phi} and i1 and o1'

psi = [phi, phi1, phi2]
table = ttg.Truths(atoms, psi)
#print(table.as_tabulate(index=False))
tbl_df = table.as_pandas()
tbl_df.columns = ['i0', 'i1', 'i2', 'i3', 'a1', 'a2', 'a3', 'o0', 'o1', 'o2', 'o3', 'c', 'phi','phi1','phi2']
```
